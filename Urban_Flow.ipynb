#Normalizacion de Fechas

````python

df["fecha"] = pd.to_datetime(df["fecha"], dayfirst=True, errors="coerce")


df["fecha"] = df["fecha"].fillna(pd.to_datetime("1932-01-01"))


print("\nFechas normalizadas:")
print(df["fecha"].dt.strftime("%Y-%m-%d").head(10))

print("\nCantidad de registros por mes:")
print(df["fecha"].dt.month.value_counts().sort_index())
````

#Normalizacion de Horas

````python
df["hora"] = pd.to_datetime(df["hora"], format="%H:%M", errors="coerce")

df["hora"] = df["hora"].fillna(pd.to_datetime("00:00", format="%H:%M"))

df["hora"] = df["hora"].dt.strftime("%H:%M")

df["hora"].head(10)
````

#Normalizacion de Ubicaciones

````python
df["ubicacion"] = (
    df["ubicacion"]
    .astype(str)
    .str.replace(r"[^a-zA-Z0-9\s]", "", regex=True)
    .str.upper()
    .str.strip()
)

df["ubicacion"].head(10)
````

#Limpieza y Normalizacion de patentes

````python
df["patente"] = (
    df["patente"]
    .astype(str)
    .str.replace(r"[^a-zA-Z0-9]", "", regex=True)
    .str.upper()
    .str.strip()
)

df["patente"] = df["patente"].replace("", pd.NA)

df["patente"].tail(10)
````

#Eliminar filas con valores relevantes vacios

````python
columnas_observadas = [
    "fecha",
    "hora",
    "ubicacion",
    "patente",
    "velocidad_registrada",
    "velocidad_maxima"
]

filas_antes = len(df)

df = df.dropna(subset=columnas_observadas)

filas_despues = len(df)

filas_eliminadas = filas_antes - filas_despues

print(f"Se eliminaron {filas_eliminadas} filas.")
print("Las columnas observadas son:")
for columna in columnas_observadas:
    print(f" - {columna}")
````

#Detectar y eliminar Outliers

````python
columnas_outliers = [
    "velocidad_registrada",
    "velocidad_maxima"
]

filas_antes = len(df)

for columna in columnas_outliers:
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    df = df[
        (df[columna] >= limite_inferior) &
        (df[columna] <= limite_superior)
    ]

filas_despues = len(df)
filas_eliminadas = filas_antes - filas_despues

print(f"Se eliminaron {filas_eliminadas} filas.")
print("Las columnas observadas son:")
for columna in columnas_outliers:
    print(f" - {columna}")
````

#Crear columna exceso_velocidad_real

````python
df["exceso_velocidad_real"] = df["velocidad_registrada"] - df["velocidad_maxima"]

df[["patente", "exceso_velocidad_real"]].head(10)
````

#Crear columna exceso_velocidad

````python
df["exceso_velocidad"] = df["velocidad_registrada"] - (df["velocidad_maxima"] * 1.05)

df[["patente", "exceso_velocidad"]].head(10)
````

#Eliminar filas que no cometieron infracción

````python
filas_antes = len(df)

df = df[df["exceso_velocidad"] > 0]

filas_despues = len(df)

filas_eliminadas = filas_antes - filas_despues

print(f"Se eliminaron {filas_eliminadas} filas.")
````

#Guardar dataset limpio

````python
ruta_interim = "urban_flow/data/interim/speeding_fines.csv"

df.to_csv(ruta_interim, index=False)

print("Dataset limpio guardado en:", ruta_interim)
````

Definir una clase FineAnalyzer que cumpla con las siguientes funcionalidades:

  - Ranking de los horarios donde se producen más multas (top 5). Debe retornar un DataFrame ordenado de mayor a menor y el índice debe comenzar desde 1, con las columnas de hora y cantidad.


  - Exceso promedio de velocidad. Debe retorna un flotante.

  - Exceso real promedio de velocidad . Debe retorna un flotante.

  - Contabilizar la cantidad de multas por cada  ubicación. Debe retornar un DataFrame ordenado por ubicación mostrando la ubicación y la cantidad.

#Definición de la clase

````python
class FineAnalyzer:
    def __init__(self, dataframe):
        self.__df = dataframe.copy()

    def ranking_patentes_mas_multadas(self):
        ranking = (
            self.__df["patente"]
            .value_counts()
            .head(5)
            .reset_index()
        )

        ranking.columns = ["patente", "cantidad"]
        ranking.index = range(1, len(ranking) + 1)

        return ranking

    def ranking_horarios_mas_multas(self):
        ranking = (
            self.__df["hora"]
            .value_counts()
            .head(5)
            .reset_index()
        )

        ranking.columns = ["hora", "cantidad"]
        ranking.index = range(1, len(ranking) + 1)

        return ranking

    def exceso_promedio_velocidad(self):
        return float(self.__df["exceso_velocidad"].mean())

    def exceso_real_promedio_velocidad(self):
        return float(self.__df["exceso_velocidad_real"].mean())

    def multas_por_ubicacion(self):
        multas = (
            self.__df["ubicacion"]
            .value_counts()
            .reset_index()
        )

        multas.columns = ["ubicacion", "cantidad"]
        multas = multas.sort_values("ubicacion")
        multas.index = range(1, len(multas) + 1)

        return multas
````
