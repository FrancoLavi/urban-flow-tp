#Normalizacion de Fechas

````python
# Conversión de la columna 'fecha' a tipo datetime
df["fecha"] = pd.to_datetime(df["fecha"], dayfirst=True, errors="coerce")

# Reemplazo de fechas inválidas por una fecha por defecto (1932-01-01)
df["fecha"] = df["fecha"].fillna(pd.to_datetime("1932-01-01"))

# Visualización de las primeras 10 fechas en formato YYYY-MM-DD
print("\nFechas normalizadas:")
print(df["fecha"].dt.strftime("%Y-%m-%d").head(10))

# Verificación de la distribución de registros por mes
print("\nCantidad de registros por mes:")
print(df["fecha"].dt.month.value_counts().sort_index())
````

#Normalizacion de Horas

````python
# Conversión de la columna 'hora' a tipo datetime en formato HH:MM (24 horas)
# Los valores inválidos se convierten en NaT
df["hora"] = pd.to_datetime(df["hora"], format="%H:%M", errors="coerce")

# Reemplazo de horas inválidas por 00:00 según la consigna
df["hora"] = df["hora"].fillna(
  pd.to_datetime("00:00", format="%H:%M")
)

# Conversión a formato string HH:MM para normalización final
df["hora"] = df["hora"].dt.strftime("%H:%M")

# Visualización de las primeras 10 horas normalizadas
df["hora"].head(10)
````

#Normalizacion de Ubicaciones

````python
# Normalización de la columna 'ubicacion'
# Se convierte a string para evitar errores con valores no textuales
# Se eliminan caracteres especiales
# Se convierte todo a mayúsculas para unificar formato
# Se eliminan espacios en blanco al inicio y al final
df["ubicacion"] = (
  df["ubicacion"]
  .astype(str)
  .str.replace(r"[^a-zA-Z0-9\s]", "", regex=True)
  .str.upper()
  .str.strip()
)

# Visualización de las primeras 10 ubicaciones normalizadas
df["ubicacion"].head(10)
````

#Limpieza y Normalizacion de patentes

````python
# Se convierte a string para evitar errores con valores no textuales
# Se eliminan caracteres especiales (solo se permiten letras y números)
# Se convierte a mayúsculas para unificar formato
# Se eliminan espacios en blanco al inicio y al final
df["patente"] = (
  df["patente"]
  .astype(str)
  .str.replace(r"[^a-zA-Z0-9]", "", regex=True)
  .str.upper()
  .str.strip()
)

# Reemplazo de valores vacíos por pd.NA (valor faltante)
df["patente"] = df["patente"].replace("", pd.NA)

# Visualización de las últimas 10 patentes normalizadas
df["patente"].tail(10)
````

#Eliminar filas con valores relevantes vacios

````python
# Definición de las columnas relevantes para el análisis de multas
# Estas columnas no deben contener valores nulos
columnas_observadas = [
  "fecha",
  "hora",
  "ubicacion",
  "patente",
  "velocidad_registrada",
  "velocidad_maxima"
]

# Cantidad de registros antes de la limpieza
filas_antes = len(df)

# Eliminación de filas que tengan valores nulos en las columnas relevantes
df = df.dropna(subset=columnas_observadas)

# Cantidad de registros después de la limpieza
filas_despues = len(df)

# Cálculo de la cantidad de filas eliminadas
filas_eliminadas = filas_antes - filas_despues

# Resultado de la limpieza
print(f"Se eliminaron {filas_eliminadas} filas.")
print("Las columnas observadas son:")
for columna in columnas_observadas:
  print(f" - {columna}")
````

#Detectar y eliminar Outliers

````python
# Columnas sobre las cuales se realizará la detección de outliers
columnas_outliers = [
  "velocidad_registrada",
  "velocidad_maxima"
]

# Cantidad de registros antes del filtrado
filas_antes = len(df)

# Eliminación de outliers utilizando el método del rango intercuartílico (IQR)
for columna in columnas_outliers:
  # Cálculo de los cuartiles
  Q1 = df[columna].quantile(0.25)
  Q3 = df[columna].quantile(0.75)

  # Cálculo del rango intercuartílico
  IQR = Q3 - Q1

  # Definición de límites inferior y superior
  limite_inferior = Q1 - 1.5 * IQR
  limite_superior = Q3 + 1.5 * IQR

  # Filtrado de valores dentro de los límites definidos
  df = df[
    (df[columna] >= limite_inferior) &
    (df[columna] <= limite_superior)
  ]

# Cantidad de registros después del filtrado
filas_despues = len(df)

# Cálculo de filas eliminadas
filas_eliminadas = filas_antes - filas_despues

# Resultado del proceso
print(f"Se eliminaron {filas_eliminadas} filas.")
print("Las columnas observadas son:")
for columna in columnas_outliers:
  print(f" - {columna}")
````

#Crear columna exceso_velocidad_real

````python
# Cálculo del exceso de velocidad real
# Se obtiene como la diferencia entre la velocidad registrada y la máxima permitida
df["exceso_velocidad_real"] = (
  df["velocidad_registrada"] - df["velocidad_maxima"]
)

# Visualización de las primeras 10 patentes con su exceso real
df[["patente", "exceso_velocidad_real"]].head(10)
````

#Crear columna exceso_velocidad

````python
# Cálculo del exceso de velocidad considerando tolerancia del 5%
# Se resta a la velocidad registrada el límite máximo incrementado en un 5%
df["exceso_velocidad"] = (
  df["velocidad_registrada"] -
  (df["velocidad_maxima"] * 1.05)
)

# Visualización de las primeras 10 patentes con su exceso de velocidad
df[["patente", "exceso_velocidad"]].head(10)
````

#Eliminar filas que no cometieron infracción

````python
# Cantidad de registros antes del filtrado
filas_antes = len(df)

# Eliminación de registros que no representan infracciones
# Se conservan únicamente aquellos donde el exceso de velocidad es mayor a 0
df = df[df["exceso_velocidad"] > 0]

# Cantidad de registros después del filtrado
filas_despues = len(df)

# Cálculo de la cantidad de filas eliminadas
filas_eliminadas = filas_antes - filas_despues

# Resultado del proceso
print(f"Se eliminaron {filas_eliminadas} filas.")
````

Definir una clase FineAnalyzer que cumpla con las siguientes funcionalidades:

  - Ranking de los horarios donde se producen más multas (top 5). Debe retornar un DataFrame ordenado de mayor a menor y el índice debe comenzar desde 1, con las columnas de hora y cantidad.


  - Exceso promedio de velocidad. Debe retorna un flotante.

  - Exceso real promedio de velocidad . Debe retorna un flotante.

  - Contabilizar la cantidad de multas por cada  ubicación. Debe retornar un DataFrame ordenado por ubicación mostrando la ubicación y la cantidad.

#Definición de la clase

````python
class FineAnalyzer:
  """
  Clase encargada de analizar el dataset de infracciones de velocidad.
  Encapsula el DataFrame y provee métodos de análisis.
  """

  def __init__(self, dataframe: pd.DataFrame) -> None:
    # Copia del DataFrame para evitar modificaciones externas
    self.__df: pd.DataFrame = dataframe.copy()

  def ranking_patentes_mas_multadas(self) -> pd.DataFrame:
    """
    Retorna el top 5 de patentes con mayor cantidad de infracciones.
    """
    ranking = (
      self.__df["patente"]
      .value_counts()
      .head(5)
      .reset_index()
    )

    ranking.columns = ["patente", "cantidad"]
    ranking.index = range(1, len(ranking) + 1)

    return ranking

  def ranking_horarios_mas_multas(self) -> pd.DataFrame:
    """
    Retorna el top 5 de horarios con mayor cantidad de infracciones.
    """
    ranking = (
      self.__df["hora"]
      .value_counts()
      .head(5)
      .reset_index()
    )

    ranking.columns = ["hora", "cantidad"]
    ranking.index = range(1, len(ranking) + 1)

    return ranking

  def exceso_promedio_velocidad(self) -> float:
    """
    Retorna el exceso promedio de velocidad con tolerancia.
    """
    return float(self.__df["exceso_velocidad"].mean())

  def exceso_real_promedio_velocidad(self) -> float:
    """
    Retorna el exceso promedio de velocidad sin tolerancia.
    """
    return float(self.__df["exceso_velocidad_real"].mean())

  def multas_por_ubicacion(self) -> pd.DataFrame:
    """
    Retorna la cantidad de infracciones por ubicación ordenadas alfabéticamente.
    """
    multas = (
      self.__df["ubicacion"]
      .value_counts()
      .reset_index()
    )

    multas.columns = ["ubicacion", "cantidad"]
    multas = multas.sort_values("ubicacion")
    multas.index = range(1, len(multas) + 1)

    return multas
````

````python
# Se excluye la fecha 1932-01-01 (dato imputado)
df_fecha = df[df["fecha"] != pd.to_datetime("1932-01-01")].copy()

# Exceso promedio por fecha
excesos_por_fecha = (
  df_fecha.groupby("fecha")["exceso_velocidad"]
  .mean()
  .sort_index()
)

# Aplicación de promedio móvil de 7 días
excesos_suavizado = excesos_por_fecha.rolling(window=7).mean()

plt.figure(figsize=(12, 5))

# Línea original (transparente)
plt.plot(
  excesos_por_fecha.index,
  excesos_por_fecha.values,
  alpha=0.3,
  label="Original"
)

# Línea suavizada
plt.plot(
  excesos_suavizado.index,
  excesos_suavizado.values,
  color="red",
  label="Promedio móvil (7 días)"
)

plt.title("Exceso promedio de velocidad por fecha")
plt.xlabel("Fecha")
plt.ylabel("Exceso promedio de velocidad")

plt.legend()
plt.tight_layout()

plt.savefig("urban_flow/data/interim/plots/date.jpg")
plt.show()
````
